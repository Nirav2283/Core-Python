### 04) WAP to demonstrate the use of Pie chart.

```python
dept = ['a' , 'b' , 'c'  , 'd']
student = [4000 , 2500 , 1500 , 1200]
e = [ 0 ,0 , 0  , 0.1]
plt.pie(student , labels = dept , explode = e , shadow = True , autopct =  "%%1.1f%%")
plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

```python
month = ['Jan' , 'Feb' , 'March' , 'April' ,  'May' , 'June' , 'July']
visitor = [11000 , 12000, 8000 , 7000 , 13000 , 4000 , 1000]
w = [0.1 , 0.2 , 0.3 , 0.4 , 0.9 , 0.10 , 0.1]
c = ['b' , 'r' , 'g' , 'c' , 'm']
plt.bar(month , visitor , width = w , color = c)
plt.show()
```

### 07) WAP to demonstrate the use of Histogram.

```python
import random as r
age = [r.randint(1,50) for i in range(100)]
plt.hist(age , edgecolor = 'k' , histtype = 'stepfilled' , align = 'mid')
plt.show()
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python

dept = ['a', 'b', 'c', 'd']
student = [4000, 2500, 1500, 1200]
plt.bar(dept, student, color='blue')
for i, value in enumerate(student):
    plt.text(i, value + 50, str(value), ha='center', va='bottom')

plt.title('Student Count by Department')
plt.xlabel('Department')
plt.ylabel('Number of Students')
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import matplotlib.pyplot as plt
import random as r
age1 = [r.randint(1, 50) for i in range(100)]
age2 = [r.randint(10, 60) for i in range(100)]
age3 = [r.randint(20, 70) for i in range(100)]

colors = ['r', 'b', 'g']

plt.hist([age1, age2, age3], histtype='stepfilled', align='mid', color=colors, label=['Group 1', 'Group 2', 'Group 3'])
plt.legend()
plt.show()
```
